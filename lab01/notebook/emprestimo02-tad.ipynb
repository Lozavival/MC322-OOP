# Transformando Empréstimo em um TAD

Considere a resolução para múltiplos empréstimos.

```c
#include <stdio.h>

typedef struct {
    float s;
    int   n;
    float j;
    int   corrente;
    float p;
} Emprestimo;

Emprestimo novoEmprestimo(float s, int n, float j) {
    Emprestimo umEmprestimo;
    umEmprestimo.s = s;
    umEmprestimo.n = n;
    umEmprestimo.j = j;
    umEmprestimo.corrente = 1;
    umEmprestimo.p = s;
    return umEmprestimo;
}

float proximaParcela(Emprestimo *umEmprestimo) {
    float retorno = umEmprestimo->p;
    if (umEmprestimo->corrente < umEmprestimo->n)
        umEmprestimo->p = umEmprestimo->p + (umEmprestimo->p * (umEmprestimo->j/100));
    else
        umEmprestimo->p = 0;
    (umEmprestimo->corrente)++;
    return retorno;
}

int main() {
    Emprestimo emprestimo1,
               emprestimo2;
    
    emprestimo1 = novoEmprestimo(200, 5, 1);
    emprestimo2 = novoEmprestimo(500, 7, 2);
    
    float p1 = proximaParcela(&emprestimo1),
          p2 = proximaParcela(&emprestimo2);
    
    int i = 1;
    while (p1 > 0 || p2 > 0) {
        if (p1 > 0)
            printf("Emprestimo 1: parcela %d eh %3.2f\n", i, p1);
        if (p2 > 0)
            printf("Emprestimo 2: parcela %d eh %3.2f\n", i, p2);
        p1 = proximaParcela(&emprestimo1);
        p2 = proximaParcela(&emprestimo2);
        i++;
    }   
}
```

# Exercício

Modifique este código (ou um código que você mesmo tenha feito) para que a estrutura (struct) possa representar um Tipo Abstrato de Dados (TAD), em que ela contenha não apenas os valores de estado,como também operações executadas sobre eles. As operações (funções em C) também devem fazer parte do struct do mesmo modo que os campos de dados.

```c
#include <stdio.h>

typedef struct Emprestimo {
    double s, j, valor_atual;
    int n, parcela_corrente;
    double (*proximaParcela)(struct Emprestimo*);
} Emprestimo;

typedef struct {
    Emprestimo emprestimos[101];
    int total_emprestimos, emprestimos_faltantes;
} Controle;

double proximaParcela(struct Emprestimo *);

void cria_emprestimo(Controle *controle, double s, int n, double j) {
    Emprestimo novo_emprestimo;
    
    novo_emprestimo.s = novo_emprestimo.valor_atual = s;
    novo_emprestimo.n = n;
    novo_emprestimo.j = j;
    novo_emprestimo.parcela_corrente = 1;
    
    controle->emprestimos[controle->total_emprestimos + 1] = novo_emprestimo;
    controle->total_emprestimos++;
    controle->emprestimos_faltantes++;
}

double proximaParcela(struct Emprestimo *emprestimo) {
    if (emprestimo->parcela_corrente > emprestimo->n)
        emprestimo->valor_atual = 0;
    else if (emprestimo->parcela_corrente > 1)
        emprestimo->valor_atual = emprestimo->valor_atual * (1 + emprestimo->j/100);
    emprestimo->parcela_corrente++;
    return emprestimo->valor_atual;
}

int main() {
    Controle controle;
    Emprestimo *emprestimo_atual;
    
    controle.total_emprestimos = controle.emprestimos_faltantes = 0;
    
    cria_emprestimo(&controle, 200, 5, 1);
    cria_emprestimo(&controle, 500, 7, 2);
    
    while (controle.emprestimos_faltantes)
        for (int i = 1; i < controle.total_emprestimos+1; i++) {
            emprestimo_atual = &(controle.emprestimos[i]);
            emprestimo_atual->proximaParcela(emprestimo_atual);
            if (emprestimo_atual->valor_atual)
                printf("Emprestimo %d: parcela %d eh %.2lf\n", i, emprestimo_atual->parcela_corrente-1, emprestimo_atual->valor_atual);
            else
                controle.emprestimos_faltantes--;
        }
}
```
